### 1. Exploring Classifiers to Predict Fatalism and Using a Mixed-data Clustering Algorithm to Segment Respondents of the HINTS Survey

### Intro

Non-adherence costs the US healthcare system an estimate of '$100B-$300B' per year. Non-adherence is associated with fatalism, the holding of deep-seated beliefs that a disease or condition is predetermined, fatal, and beyond one's control.

One way health costs could potentially be reduced is by identifying patients who are fatalistic and counselling with them regarding their beliefs. However, it is not very commom for fatalistic patients to reveal their views to medical providers. Therefore, it would be helpful to develop ML models to see if these views/patients can be predicted using information that may be available at the clinical point of care, and in the future, integrate these predictions into the EHR so they are available to the medical team.


Sources: https://www.pharmacytimes.com/view/what-is-the-true-cost-of-medication-nonadherence

-Sinky TH, Faith J, Lindly O, Thorburn S. Cancer Fatalism and Preferred Sources of Cancer Information: an Assessment Using 2012 HINTS Data. J Cancer Educ. 2018 Feb;33(1):231-237. doi: 10.1007/s13187-016-1115-1. PMID: 27650861.

-De Jesus M, Miller EB (2014) Fatalistic beliefs or structural factors impeding screening?: examining “real” barriers to breast cancer screening among Latina immigrants. Health Care for Women Int. doi:10.1080/07399332.2014.973496

-Maria De Jesus & Eva B. Miller (2015) Examining Breast Cancer Screening Barriers Among Central American and Mexican Immigrant Women: Fatalistic Beliefs or Structural Factors?, Health Care for Women International, 36:5, 593-607, DOI: doi:10.1080/07399332.2014.973496



###  Dataset Overview

The dataset comes from the Health Information National Trends Survey- 2012 conducted by the National Cancer Institute. The purpose of the survey was to collect data on how Americans consume information related to cancer. 
<br> </br> 
The data has 3630 respondents and 357 features. The data can be accessed at https://hints.cancer.gov/data/download-data.aspx.

The feature of inteterests for the classification experiments are: 'PreventNotPossible'? (belief that it's not possiblie to prevent cancer) and 'EverythingCauseCancer' (belief that everything causes cancer). I combined these variables into one final outcome variable: 'fatalview', if a respondent agreed to any of these questions.

```python
from scipy.stats import loguniform
from statistics import mean
from scipy import stats
#import scikit_posthocs
import warnings

np.random.seed(0)
target_names = ['class 0', 'class 1'] 
scoresTests=[]

for i in range(0,3):
    print('Classifier_'+str(i)+' --> ',classifiers[i])
    
    model= RandomizedSearchCV(classifiers[i], paramatersClassifiers[i], scoring='balanced_accuracy', cv=rsFolds, random_state=1)
    model.fit(X_train,y_train)
    
    print('Best Score: %s' % model.best_score_)
    print('Best Hyperparameters: %s' % model.best_params_)
    
    y_pred=model.predict(X_train)
    
    print(classification_report(y_train, y_pred, target_names=target_names))
    
    scorePerClassifier=[]
    
    for j in range(0, len(test_splits)):
        y_test_pred=model.predict(test_splits[j][0])
        scorePerClassifier.append(f1_score(test_splits[j][1], y_test_pred, average='micro'))
    
    print (mean(scorePerClassifier))
    scoresTests.append(scorePerClassifier)      
```

```python
# Convert dataframe to matrix
dfMatrix = Data.to_numpy()
```

```python
#Elbow method to detect number of K
from kmodes.kprototypes import KPrototypes

cost = []
for cluster in range(1, 10):
    try:
        kprototype = KPrototypes(n_jobs = -1, n_clusters = cluster, init = 'Huang', random_state = 0)
        kprototype.fit_predict(dfMatrix, categorical = catColumnsPos)
        cost.append(kprototype.cost_)
        print('Cluster initiation: {}'.format(cluster))
    except:
        break

plt.plot(cost)
plt.xlabel('K')
plt.ylabel('cost')
plt.show
```

```python
# cost (sum distance): confirm visual clue of elbow plot
# KneeLocator to detect elbows 
import kneed
from kneed import KneeLocator
cost_knee_c3 = KneeLocator(
        x=range(1,10), 
        y=cost, 
        S=0.1, curve="convex", direction="decreasing", online=True)

K_cost_c3 = cost_knee_c3.elbow   
print("elbow at k =", f'{K_cost_c3:.0f} clusters')
```

```python
#Kprototype model with 3 clusters
kprototype = KPrototypes(n_jobs = -1, n_clusters = 3, init = 'Cao', random_state = 0)
Data['clusters']= kprototype.fit_predict(dfMatrix, categorical = catColumnsPos)
```

### 8. Discussion

The main objective of this project was to use the HINTS data to develop and compare classification models to identify people that may hold fatalistic views with respect to cancer in this case. After the experiment, logistic regression was the best performing model with an averaged balanced accuracy of 79%, however, due to high imbalance of the positive class , 'fatalview' = yes, there is a extremely low recall. In contrast, XGB, which came in second for averaged balanced accuracy, had much better recall at 65%.

A secondary aim was to find groups or "segment" the people that took the HINTS survey using the k-prototype algorithm, which is able to cluster dataframes with mixed data types (numerical and categorical). The results indicated there are three 3 clusters in the data of clearly different age and income groups.

A future direction for this project is to use SMOTE and rerun the models.


### Limitations
This project has a number of limitations. First of all, during data preprocessing, I opted for the imputation of categorical variables with mode and mean, which could have introduced bias. Since the target variable also had some NaN's, I also dropped ~200 records that had a NaN in the target variable 'fatalview.' Another limitation was that the data had several features with imbalance, including the target variable 'fatalview'. Although, I tried to address this by using repeated stratified k-fold cross validation, and when evaluating the models using the balanced accuracy. Finally, one more limitation explained above was the sparseness of the data and what this entailed for the clustering algorithms and lack of cluster differentiation.
